```python
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']

df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar', title = 'Purchase Distribution in Age')
```

```python
list = []

for i in df['Age'].unique():
  list.append([i, df[df['Age'] == i]['Product_ID'].nunique()])

data = pd.DataFrame(list, columns = ['Age', 'Products'])
```

```python
data.plot.bar(x = 'Age', figsize = (12, 6))
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind = 'pie', title = 'Amount spend in Age', autopct = '%.2f')
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind = 'pie', title = 'Amount spend in Age', autopct = '%.2f')
```

```python
df.groupby('Marital_Status').size().plot(kind = 'pie', autopct = '%.2f')
```

```python
df.groupby('Gender').size().plot(kind = 'pie', autopct = '%.2f')
```
